```python
import pandas as pd
rds=pd.read_csv('/content/drive/MyDrive/PDS/Question1/Raw_Data/RAW.csv')
rds
```

```python
rds.columns=['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)','Frailty']
rds.columns
rds['Frailty'].replace(['Y', 'N'],
                        [0, 1], inplace=True)
```

```python

clean_data=rds.dropna()
clean_data
```

```python
clean_data.to_csv('/content/drive/MyDrive/PDS/Question1/Clean_Data/Clean_Data.csv')
print('successful!')
```

```python
import matplotlib.pyplot as plt
v= plt.scatter(clean_data['Height(in)'], clean_data['Weight(lb)'], alpha=0.5)
plt.xlabel('Height (in)')
plt.ylabel('Weight (lb)')
plt.title('Weight vs Height')
plt.show()
fig = v.get_figure()
fig.savefig("/content/drive/MyDrive/PDS/Question1/Results/weightvsheight.png")
```

```python
import seaborn as sns
v=sns.boxplot(x='Frailty', y='Grip_strength(kg)', data=clean_data)
plt.title('Grip Strength & Frailty')
plt.show()
fig = v.get_figure()
fig.savefig("/content/drive/MyDrive/PDS/Question1/Results/gripstrengthbyfrailty.png")
```

```python
import seaborn as sns
v=sns.countplot(x='Frailty', data=clean_data)
plt.title('Frailty Distribution')
plt.show()
fig = v.get_figure()
fig.savefig("/content/drive/MyDrive/PDS/Question1/Results/frailtydistribution.png")
```
